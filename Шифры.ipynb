# Шифр Цезаря

```python
# - шифр Цезаря шифрует по буквам алфавита.
def caesar_cipher(text, shift, language='english'):
  """Шифрует или дешифрует текст шифром Цезаря.

  Args:
    text: Текст для шифрования/дешифровки.
    shift: Сдвиг для шифра.
    language: Язык текста ('english' или 'russian').

  Returns:
    Зашифрованный/дешифрованный текст.
  """
  result = ''
  if language == 'english':
    alphabet = 'abcdefghijklmnopqrstuvwxyz'
  elif language == 'russian':
    alphabet = 'абвгдеёжзийклмнопрстуфхцчшщъыьэюя'
  else:
    return 'Неверный язык'

  for char in text:
    if char.lower() in alphabet:
      char_index = alphabet.find(char.lower())
      if char.isupper():
        new_index = (char_index + shift) % len(alphabet)
        result += alphabet[new_index].upper()
      else:
        new_index = (char_index + shift) % len(alphabet)
        result += alphabet[new_index]
    else:
      result += char

  return result


def main():
  while True:
    print("\nВыберите действие:")
    print("1. Шифрование")
    print("2. Дешифрование")
    print("3. Выход")

    choice = input("Введите номер действия: ")

    if choice == '1':
      text = input("Введите текст для шифрования: ")
      shift = int(input("Введите сдвиг: "))
      print("Выберите язык:")
      print("1. Английский")
      print("2. Русский")
      lang_choice = input("Введите номер языка: ")
      if lang_choice == '1':
        language = 'english'
      elif lang_choice == '2':
        language = 'russian'
      else:
        print("Неверный выбор языка.")
        continue
      encrypted_text = caesar_cipher(text, shift, language)
      print("Зашифрованный текст:", encrypted_text)
    elif choice == '2':
      text = input("Введите текст для дешифровки: ")
      shift = int(input("Введите сдвиг: "))
      print("Выберите язык:")
      print("1. Английский")
      print("2. Русский")
      lang_choice = input("Введите номер языка: ")
      if lang_choice == '1':
        language = 'english'
      elif lang_choice == '2':
        language = 'russian'
      else:
        print("Неверный выбор языка.")
        continue
      decrypted_text = caesar_cipher(text, -shift, language)
      print("Дешифрованный текст:", decrypted_text)
    elif choice == '3':
      break
    else:
      print("Неверный выбор.")

if __name__ == "__main__":
  main()
```

# Шифр Виженера

```python
# Виженера — шифрует с помощью ключа.

def vigenere_cipher(text, key, mode='encrypt', language='english'):
  """Шифрует или дешифрует текст шифром Виженера.

  Args:
    text: Текст для шифрования/дешифровки.
    key: Ключ для шифра.
    mode: Режим работы ('encrypt' или 'decrypt').
    language: Язык текста ('english' или 'russian').

  Returns:
    Зашифрованный/дешифрованный текст.
  """
  result = ''
  if language == 'english':
    alphabet = 'abcdefghijklmnopqrstuvwxyz'
  elif language == 'russian':
    alphabet = 'абвгдеёжзийклмнопрстуфхцчшщъыьэюя'
  else:
    return 'Неверный язык'

  key_index = 0
  for char in text:
    if char.lower() in alphabet:
      char_index = alphabet.find(char.lower())
      key_char = key[key_index % len(key)].lower()
      key_char_index = alphabet.find(key_char)

      if mode == 'encrypt':
        new_index = (char_index + key_char_index) % len(alphabet)
      elif mode == 'decrypt':
        new_index = (char_index - key_char_index) % len(alphabet)
      else:
        return 'Неверный режим'

      if char.isupper():
        result += alphabet[new_index].upper()
      else:
        result += alphabet[new_index]

      key_index += 1
    else:
      result += char

  return result


def main():
  while True:
    print("\nВыберите действие:")
    print("1. Шифрование")
    print("2. Дешифрование")
    print("3. Выход")

    choice = input("Введите номер действия: ")

    if choice == '1':
      text = input("Введите текст для шифрования: ")
      key = input("Введите ключ: ")
      print("Выберите язык:")
      print("1. Английский")
      print("2. Русский")
      lang_choice = input("Введите номер языка: ")
      if lang_choice == '1':
        language = 'english'
      elif lang_choice == '2':
        language = 'russian'
      else:
        print("Неверный выбор языка.")
        continue
      encrypted_text = vigenere_cipher(text, key, 'encrypt', language)
      print("Зашифрованный текст:", encrypted_text)
    elif choice == '2':
      text = input("Введите текст для дешифровки: ")
      key = input("Введите ключ: ")
      print("Выберите язык:")
      print("1. Английский")
      print("2. Русский")
      lang_choice = input("Введите номер языка: ")
      if lang_choice == '1':
        language = 'english'
      elif lang_choice == '2':
        language = 'russian'
      else:
        print("Неверный выбор языка.")
        continue
      decrypted_text = vigenere_cipher(text, key, 'decrypt', language)
      print("Дешифрованный текст:", decrypted_text)
    elif choice == '3':
      break
    else:
      print("Неверный выбор.")

if __name__ == "__main__":
  main()
```

# Шифр Атбаша

```python

# Атбаш — шифрует в обратном порядке.

def atbash_cipher(text, language='english'):
  """Шифрует или дешифрует текст шифром Атбаш.

  Args:
    text: Текст для шифрования/дешифровки.
    language: Язык текста ('english' или 'russian').

  Returns:
    Зашифрованный/дешифрованный текст.
  """
  result = ''
  if language == 'english':
    alphabet = 'abcdefghijklmnopqrstuvwxyz'
  elif language == 'russian':
    alphabet = 'абвгдеёжзийклмнопрстуфхцчшщъыьэюя'
  else:
    return 'Неверный язык'

  for char in text:
    if char.lower() in alphabet:
      char_index = alphabet.find(char.lower())
      new_index = len(alphabet) - 1 - char_index
      if char.isupper():
        result += alphabet[new_index].upper()
      else:
        result += alphabet[new_index]
    else:
      result += char

  return result


def main():
  while True:
    print("\nВыберите действие:")
    print("1. Шифрование")
    print("2. Дешифрование")
    print("3. Выход")

    choice = input("Введите номер действия: ")

    if choice == '1':
      text = input("Введите текст для шифрования: ")
      print("Выберите язык:")
      print("1. Английский")
      print("2. Русский")
      lang_choice = input("Введите номер языка: ")
      if lang_choice == '1':
        language = 'english'
      elif lang_choice == '2':
        language = 'russian'
      else:
        print("Неверный выбор языка.")
        continue
      encrypted_text = atbash_cipher(text, language)
      print("Зашифрованный текст:", encrypted_text)
    elif choice == '2':
      text = input("Введите текст для дешифровки: ")
      print("Выберите язык:")
      print("1. Английский")
      print("2. Русский")
      lang_choice = input("Введите номер языка: ")
      if lang_choice == '1':
        language = 'english'
      elif lang_choice == '2':
        language = 'russian'
      else:
        print("Неверный выбор языка.")
        continue
      decrypted_text = atbash_cipher(text, language)
      print("Дешифрованный текст:", decrypted_text)
    elif choice == '3':
      break
    else:
      print("Неверный выбор.")


if __name__ == "__main__":
  main()
```

# Шифр Плейфера

```python
# Импорт необходимых библиотек
import numpy as np

# Создание квадрата Плейфера для английского языка
def create_playfair_square(key):
    key = key.replace('J', 'I').upper()
    alphabet = 'ABCDEFGHIKLMNOPQRSTUVWXYZ'
    key_set = set(key)
    square = []
    for char in key:
        if char not in square:
            square.append(char)
    for char in alphabet:
        if char not in key_set and char not in square:
            square.append(char)
    return np.array(square).reshape(5, 5)

# Поиск позиции буквы в квадрате Плейфера
def find_position(square, char):
    index = np.where(square == char)
    return index[0][0], index[1][0]

# Шифрование пары букв
def encrypt_pair(square, a, b):
    row_a, col_a = find_position(square, a)
    row_b, col_b = find_position(square, b)

    if row_a == row_b:
        return square[row_a, (col_a + 1) % 5], square[row_b, (col_b + 1) % 5]
    elif col_a == col_b:
        return square[(row_a + 1) % 5, col_a], square[(row_b + 1) % 5, col_b]
    else:
        return square[row_a, col_b], square[row_b, col_a]

# Дешифрование пары букв
def decrypt_pair(square, a, b):
    row_a, col_a = find_position(square, a)
    row_b, col_b = find_position(square, b)

    if row_a == row_b:
        return square[row_a, (col_a - 1) % 5], square[row_b, (col_b - 1) % 5]
    elif col_a == col_b:
        return square[(row_a - 1) % 5, col_a], square[(row_b - 1) % 5, col_b]
    else:
        return square[row_a, col_b], square[row_b, col_a]

# Шифрование текста
def encrypt(text, square):
    text = text.replace('J', 'I').upper().replace(' ', '')
    if len(text) % 2 != 0:
        text += 'X'
    encrypted_text = ''
    for i in range(0, len(text), 2):
        a, b = text[i], text[i + 1]
        if a == b:
            b = 'X'
        encrypted_text += ''.join(encrypt_pair(square, a, b))
    return encrypted_text

# Дешифрование текста
def decrypt(text, square):
    text = text.upper().replace(' ', '')
    decrypted_text = ''
    for i in range(0, len(text), 2):
        a, b = text[i], text[i + 1]
        decrypted_text += ''.join(decrypt_pair(square, a, b))
    return decrypted_text

# Ввод текста от пользователя
def get_user_input():
    return input("Введите текст: ")

# Ввод ключа от пользователя
def get_key_input():
    return input("Введите ключ для создания квадрата Плейфера: ")

# Выбор варианта (шифрование или дешифрование)
def choose_option():
    while True:
        option = input("Выберите вариант (1 - шифрование, 2 - дешифрование): ").strip()
        if option in ['1', '2']:
            return option
        else:
            print("Неверный выбор. Пожалуйста, выберите '1' или '2'.")

# Пример использования
key = get_key_input()
text = get_user_input()
option = choose_option()

square = create_playfair_square(key)

if option == '1':
    result = encrypt(text, square)
    print(f"Зашифрованный текст: {result}")
elif option == '2':
    result = decrypt(text, square)
    print(f"Расшифрованный текст: {result}")
```

# Шифр Вернама


```python
# Импорт необходимых библиотек
import random

# Функция для преобразования текста в числа
def text_to_numbers(text):
    return [ord(char) for char in text]

# Функция для преобразования чисел в текст
def numbers_to_text(numbers):
    return ''.join(chr(num) for num in numbers)

# Функция для генерации случайного ключа
def generate_key(length):
    return [random.randint(0, 255) for _ in range(length)]

# Функция для шифрования текста
def encrypt(text, key):
    text_numbers = text_to_numbers(text)
    encrypted_numbers = [text_numbers[i] ^ key[i] for i in range(len(text_numbers))]
    return encrypted_numbers

# Функция для дешифрования текста
def decrypt(encrypted_numbers, key):
    decrypted_numbers = [encrypted_numbers[i] ^ key[i] for i in range(len(encrypted_numbers))]
    return numbers_to_text(decrypted_numbers)

# Ввод текста от пользователя
def get_user_input():
    return input("Введите текст: ")

# Выбор варианта (шифрование или дешифрование)
def choose_option():
    while True:
        option = input("Выберите вариант (1 - шифрование, 2 - дешифрование): ").strip()
        if option in ['1', '2']:
            return option
        else:
            print("Неверный выбор. Пожалуйста, выберите '1' или '2'.")

# Пример использования
text = get_user_input()
option = choose_option()

key = generate_key(len(text))

if option == '1':
    encrypted_numbers = encrypt(text, key)
    print(f"Ключ: {key}")
    print(f"Зашифрованные числа: {encrypted_numbers}")
elif option == '2':
    encrypted_numbers = eval(input("Введите зашифрованные числа (в формате списка, например, [1, 2, 3]): "))
    decrypted_text = decrypt(encrypted_numbers, key)
    print(f"Расшифрованный текст: {decrypted_text}")
```

# Шифр RSA

```python
# Импорт необходимых библиотек
from cryptography.hazmat.primitives.asymmetric import rsa, padding
from cryptography.hazmat.primitives import serialization, hashes
from cryptography.hazmat.backends import default_backend

# Генерация ключей
def generate_keys():
    private_key = rsa.generate_private_key(
        public_exponent=65537,
        key_size=2048,
        backend=default_backend()
    )
    public_key = private_key.public_key()
    return private_key, public_key

# Шифрование текста
def encrypt(plaintext, public_key):
    ciphertext = public_key.encrypt(
        plaintext.encode('utf-8'),
        padding.OAEP(
            mgf=padding.MGF1(algorithm=hashes.SHA256()),
            algorithm=hashes.SHA256(),
            label=None
        )
    )
    return ciphertext

# Дешифрование текста
def decrypt(ciphertext, private_key):
    plaintext = private_key.decrypt(
        ciphertext,
        padding.OAEP(
            mgf=padding.MGF1(algorithm=hashes.SHA256()),
            algorithm=hashes.SHA256(),
            label=None
        )
    )
    return plaintext.decode('utf-8')

# Ввод текста от пользователя
def get_user_input():
    return input("Введите текст для шифрования: ")

# Пример использования
plaintext = get_user_input()
private_key, public_key = generate_keys()
ciphertext = encrypt(plaintext, public_key)
decrypted_text = decrypt(ciphertext, private_key)

print(f"Исходный текст: {plaintext}")
print(f"Зашифрованный текст: {ciphertext}")
print(f"Расшифрованный текст: {decrypted_text}")
```

```python
!pip install pycryptodome
from Crypto.Cipher import DES
from Crypto.Util.Padding import pad, unpad
import binascii

def des_encrypt(plain_text, key):
    """
    Шифрует текст с использованием алгоритма DES.

    :param plain_text: Исходный текст для шифрования.
    :param key: Ключ шифрования (должен быть 8 байт).
    :return: Зашифрованный текст в шестнадцатеричном формате.
    """
    cipher = DES.new(key.encode('utf-8'), DES.MODE_ECB)
    padded_text = pad(plain_text.encode('utf-8'), DES.block_size)
    encrypted_text = cipher.encrypt(padded_text)
    return binascii.hexlify(encrypted_text).decode('utf-8')

def des_decrypt(encrypted_text, key):
    """
    Дешифрует текст с использованием алгоритма DES.

    :param encrypted_text: Зашифрованный текст в шестнадцатеричном формате.
    :param key: Ключ шифрования (должен быть 8 байт).
    :return: Расшифрованный текст.
    """
    cipher = DES.new(key.encode('utf-8'), DES.MODE_ECB)
    encrypted_bytes = binascii.unhexlify(encrypted_text)
    decrypted_padded_text = cipher.decrypt(encrypted_bytes)
    return unpad(decrypted_padded_text, DES.block_size).decode('utf-8')

def main():
    text = input("Введите текст для шифрования (до 8 символов): ")
    key = input("Введите ключ (8 символов): ")

    # Проверяем длину ключа и текста
    if len(key) != 8:
        print("Ключ должен содержать ровно 8 символов.")
        return

    if len(text) > 8:
        print("Текст должен содержать не более 8 символов.")
        return

    # Шифрование
    encrypted = des_encrypt(text, key)
    print(f"Зашифрованный текст: {encrypted}")

    # Дешифрование
    decrypted = des_decrypt(encrypted, key)
    print(f"Расшифрованный текст: {decrypted}")

if __name__ == "__main__":
    main()
```
